```python
import matplotlib.pyplot as plt
```

```python
#Creating Data for BoxPlot

data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data)

plt.show()
```

```python
#Notch Plot
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data, notch=True)

plt.show()
```

```python
#Horizontal Box Plot 
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data, vert=False)

plt.show()
```

```python
#Notch Plot
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data, widths = 0.5)#Changing the Width 

plt.show()
```

```python
#Notch Plot
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data, labels=["Marks in Math"]) #Labels must be in list


plt.show()
```

```python
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data, labels=["Marks in Math"], 
            patch_artist=True) #Filling value

plt.show()
```

```python
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data, labels=["Marks in Math"], showmeans=True) #Filling value

plt.show()
```

```python
#Outliers 
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90, 200]

plt.boxplot(data, labels=["Marks in Math"], showmeans=True) #Filling value

plt.show()
```

```python
#Joining the outliers
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90, 200]

plt.boxplot(data, labels=["Marks in Math"], showmeans=True, whis=3.5) #Filling value

plt.show()
```

```python
#Outliers and colors 
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90, 200]

plt.boxplot(data, labels=["Marks in Math"], showmeans=True, sym='k*', 
            boxprops=dict(color = 'r')) #Colors of Box 

plt.show()
```

```python
#Outliers and colors 
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data, labels=["Marks in Math"], showmeans=True, sym='k*', 
            boxprops=dict(color = 'r'), capprops=dict(color="g")) #Colors of Cap 

plt.show()
```

```python
#colors 
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

plt.boxplot(data, labels=["Marks in Math"], showmeans=True, sym='k*', 
            boxprops=dict(color = 'r'), capprops=dict(color="g"),
            whiskerprops=dict(color = "b")) #Whisker color 

plt.show()
```

```python
#Multiple Boxplot 
data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

data1 = [11, 22, 33, 44, 44, 33, 22, 55, 65, 76, 79, 34, 45, 67, 80]

y = [data, data1]

plt.boxplot(y, showmeans=True, sym='k*', 
            boxprops=dict(color = 'r'), capprops=dict(color="g")) 

plt.show()
```

```python
#Multiple Boxplot 

data = [33, 44, 55, 55, 66, 77, 88, 55, 44, 99, 88, 20, 30, 49, 40, 90]

data1 = [11, 22, 33, 44, 44, 33, 22, 55, 65, 76, 79, 34, 45, 67, 80]

y = [data, data1]

plt.boxplot(y, labels=["Math", "Science"], showmeans=True, sym='k*', 
            boxprops=dict(color = 'r'), capprops=dict(color="g")) 

plt.ylabel("Score")
plt.title("Boxplot of Math and Science")

plt.show()
```
