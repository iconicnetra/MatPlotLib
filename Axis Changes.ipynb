```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 25, 27, 35, 40]

plt.plot(x, y)

plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 25, 27, 35, 40]

plt.plot(x, y)

plt.xticks(x) #Changes according to x

plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 25, 27, 35, 40]

plt.plot(x, y)

plt.xticks(x) #Changes according to x
plt.yticks(x) #Changes according to x

plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 25, 27, 35, 40]

plt.plot(x, y)

plt.xticks(x, labels=["A", "B", "C", "D", "E", "F"]) #Changes according to rthe labels


plt.show()
```

```python
#Working with Limits

x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 25, 27, 35, 40]

plt.plot(x, y)

# plt.xticks(x, labels=["A", "B", "C", "D", "E", "F"]) #Changes according to rthe labels

plt.xlim(0, 10)
plt.ylim(10, 50)

plt.show()
```

```python
#Working with Limits

x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 25, 27, 35, 40]

plt.plot(x, y)

# plt.xticks(x, labels=["A", "B", "C", "D", "E", "F"]) #Changes according to rthe labels

# plt.xlim(0, 10)
# plt.ylim(10, 50)

plt.axis([0, 7, 0, 40]) #Changes in axis: x-axis: 0 to 7; y-axisa: 0 to 40

plt.show()
```
