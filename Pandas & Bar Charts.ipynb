```python
data = pd.read_csv('F:\Programming\Python Programming\Data\Data.csv')
data
```

```python
data.head(3)
```

```python
#Creating a Bar Plot of Religion 

religion_count = data["Religion"].value_counts()

religion_count.plot(kind='bar')
print(religion_count)

#Labels and Titles
plt.xlabel("Religion")
plt.ylabel("No. of Respondents")
plt.title("Bar Plot of Religion")
plt.legend("Religion")
plt.xticks(rotation=45, ha = 'right')
plt.grid(True, color='k', alpha = 0.3)
plt.show()
```

```python
gender_count = data['Gender'].value_counts()

gender_count.plot(kind='barh')
plt.ylabel("Gender")
plt.xlabel("No. of Participants")
plt.title("Barplot of Gender")
plt.grid(True, c='r', alpha = 0.2) #c = color and alpha is the dimness of the visual 

plt.legend("Gender", loc = 2)

plt.show()
```

```python
#Legende Postion (Both numbers and words can be used)
# 0: 'best' (automatically choose the best location)
# 1: 'upper right'
# 2: 'upper left'
# 3: 'lower left'
# 4: 'lower right'
# 5: 'right'
# 6: 'center left'
# 7: 'center right'
# 8: 'lower center'
# 9: 'upper center'
# 10: 'center'
```
