```python
import matplotlib.pyplot as plt
```

```python
import csv
import numpy as np
from collections import Counter
```

```python
plt.style.use("ggplot")

religions = []

#Reading Data from CSV file
with open('Data.csv', 'r') as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        religions.append(row['Religion'])

#Counting the occurence in the Religion column
religion_counts = Counter(religions)

#Extract unique religions and their corresponding counts 
unique_religions = list(religion_counts.keys())
counts = list(religion_counts.values())

#Create a Bar Plot
plt.bar(unique_religions, counts, color = 'y', label="Religion")

#Add labels and titles

plt.xlabel("Religion")
plt.ylabel("Number of Respondents")
plt.title("Distribution of Religion")

plt.grid(True, color = 'k', alpha = 0.3)
plt.legend()
plt.show()
```

```python
academic = []

with open("Data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        academic.append(row["Academic_Level"])

academic_count = Counter(academic)

academic_keys = list(academic_count.keys())
academic_values = list(academic_count.values())

plt.bar(academic_keys, academic_values, color='k')

plt.xlabel("Academic Level")
plt.ylabel("No. of Respondents")
plt.title("Bar Plot of Academic Level of Students")

plt.legend("Academic Level")
plt.grid(True, color = 'g', alpha = 0.3)

plt.show()
```
