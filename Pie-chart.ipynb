```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")

slices = [60, 90]

#Creating Pie-char
plt.pie(slices)

plt.title("My Awesome Pie Chart")
plt.tight_layout()

plt.show()
```

```python
#Labels
plt.style.use("fivethirtyeight")

slices = [60, 90]

#Creating Pie-char
labels = ['Sixty', 'Ninty']

plt.pie(slices, labels=labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()

plt.show()
```

```python
#Formatting Pie-chart
plt.style.use("fivethirtyeight")

slices = [60, 90, 120, 200]

#Creating Pie-char
labels = ['A', 'B', 'C', 'D']
colors = ['blue', 'red', 'yellow', 'green']

plt.pie(slices, labels=labels, colors = colors, wedgeprops={'edgecolor':'black'}) #Adding the colors in the edge of the pie chart
#we added also colors 
plt.title("My Awesome Pie Chart")
plt.tight_layout()

plt.show()
```

```python
#Formatting Pie-chart
plt.style.use("fivethirtyeight")

slices = [60, 90, 120, 200]

#Creating Pie-char
labels = ['A', 'B', 'C', 'D']
colors = ['blue', 'red', 'yellow', 'green']

#Exploding the slice

explode = [0, 0, 0.2, 0]

plt.pie(slices, labels=labels, explode=explode,
        colors = colors, wedgeprops={'edgecolor':'black'})


plt.title("My Awesome Pie Chart")
plt.tight_layout()

plt.show()
```

```python
#Formatting Pie-chart
plt.style.use("fivethirtyeight")

slices = [60, 90, 120, 200]

#Creating Pie-char
labels = ['A', 'B', 'C', 'D']
colors = ['blue', 'red', 'yellow', 'green']

#Exploding the slice

explode = [0, 0, 0.2, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        colors = colors, wedgeprops={'edgecolor':'black'})#added shadow


plt.title("My Awesome Pie Chart")
plt.tight_layout()

plt.show()
```

```python
#Formatting Pie-chart
plt.style.use("fivethirtyeight")

slices = [60, 90, 120, 200]

#Creating Pie-char
labels = ['A', 'B', 'C', 'D']
colors = ['blue', 'red', 'yellow', 'green']

#Exploding the slice

explode = [0, 0, 0.2, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        colors = colors, startangle=90,
        wedgeprops={'edgecolor':'black'})#Added starting angle....Rotating the pie-chart


plt.title("My Awesome Pie Chart")
plt.tight_layout()

plt.show()
```

```python
#Formatting Pie-chart
plt.style.use("fivethirtyeight")

slices = [60, 90, 120, 200]

#Creating Pie-char
labels = ['A', 'B', 'C', 'D']
colors = ['grey', 'red', 'yellow', 'green']

#Exploding the slice

explode = [0, 0, 0.2, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        colors = colors, startangle=90, autopct='%1.2f%%',
        wedgeprops={'edgecolor':'black'})#Adding percentation...(.2 means 2 digits after decimal)


plt.title("My Awesome Pie Chart")
plt.tight_layout()

plt.show()
```
